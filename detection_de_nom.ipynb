```python
def noun_detector(text):

  nlp = fr_core_news_sm.load()

  nlp1 = fr_core_news_md.load()

  nlp2 = fr_core_news_lg.load()

  nlp3 = fr_dep_news_trf.load()


  doc = nlp(text)
  doc1 = nlp1(text)
  doc2 = nlp2(text)
  doc3 = nlp3(text)
  
  matcher = Matcher(nlp.vocab)

  pattern = [{"POS":"PROPN"}]
  pattern2 = [{"POS":"NOUNS"}]

  matcher.add("NOUN_1",[pattern])
  matcher.add("NOUN_2",[pattern2])


  matches = matcher(doc)
  matches1 = matcher(doc1)
  matches2 = matcher(doc2)
  matches3 = matcher(doc3)
  i=0
  j=0
  k=0
  l=0
  Noms = []
  Noms1 = []
  Noms2 = []
  Noms3 = []
  print("\n************pipeline    fr_core_news_sm ******************\n\n  ")
  for Nouns,start,end in matches:
    i+=1
    if i==1:
      print(f"le {i}er  nom extraits avec le pipeline fr_core_news_md est :  {doc[start:end]}")

    else:
      print(f"\nle {i} eme  nom extraits avec le pipeline fr_core_news_md  est :  {doc[start:end]}")

    Noms.append(doc[start:end])
  print (f"Le pipeline fr_core_news_sm  a détecté {len(matches)} noms")


  print("\n************pipeline    fr_core_news_md******************\n\n  ")
  for (Nouns1,start1,end1) in matches1:
    j+=1
    if j==1:
      print(f"\nle {j}er nom extrait avec le pipeline fr_core_news_md est :  {doc1[start1:end1]}")
    else:
      print(f"\nle {j}er nom extrait avec le pipeline fr_core_news_md est :  {doc1[start1:end1]}")   
    Noms1.append(doc1[start1:end1])
  print (f"\n Le pipeline fr_core_news_md  a détecté {len(matches1)} noms")


  print("\n************pipeline   fr_core_news_lg******************\n\n  ")
  for (Nouns2,start2,end2) in matches2:
    k+=1
    if k==1:
      print(f"\nle {k}er nom extrait avec le pipeline fr_core_news_lg est :  {doc2[start2:end2]}")
    else:
      print(f"\nle {k}e nom extrait avec le pipeline fr_core_news_lg est :  {doc2[start2:end2]}")
    Noms2.append(doc2[start2:end2])
  print (f"\n Le pipeline fr_core_news_lg  a détecté {len(matches2)} noms")


  print("\n************pipeline  fr_dep_news_trf****************** \n\n ")
  for (Nouns3,start3,end3) in matches3:
    l+=1
    if l==1:
      print(f"\n le {l}er nom extrait avec le pipeline fr_core_news_lg est :  {doc3[start3:end3]}")
    else:
      print(f"\n le {l}e nom extrait avec le pipeline fr_core_news_lg est :  {doc3[start3:end3]}")   
    Noms3.append(doc3[start3:end3])
  print (f"\n Le pipeline pipeline  fr_dep_news_trf  a détecté {len(matches3)} noms")
   
  



 


  return Noms,Noms1,Noms2,Noms3


```

```python
text = " Je m'appelle du  Vincent MEliga Naga  WAYNE FRANCK Bruno Wambo et je travaille à ESOAF    grace à mon "
```

```python
print(noun_detector(text))
```
